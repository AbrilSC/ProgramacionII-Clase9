Descripción del Problema:
    Escribir un programa en Python que administre una lista de tareas. El programa debe permitir al usuario agregar, eliminar y listar tareas. Cada tarea tendrá un nombre, una descripción y una prioridad (alta, media, baja). Además, habrá diferentes tipos de tareas: TareaGeneral y TareaConFecha, cada una con características adicionales.

Requisitos del Código:
    1. El programa debe permitir agregar una nueva tarea.
    2. El programa debe permitir eliminar una tarea existente.
    3. El programa debe permitir listar todas las tareas con sus detalles.
    4. El programa debe estar diseñado utilizando POO e incluir herencia.

```python
# ******** Defino super clase tarea ********
class Tarea:
    #Constructor de la clase
    def __init__(self, nombre, descripcion, prioridad):
        self.nombre = nombre
        self.descripcion = descripcion
        self.prioridad = prioridad

    #Muestra los detalles de la tarea
    def mostrarTarea(self):
        print(f"Nombre: {self.nombre}")
        print(f"Descripcion: {self.descripcion}")
        print(f"Prioridad: {self.prioridad}")
        print("..................")  
        print()

# ******** Defino subclase Tarea General hereda de Tarea ********
class TareaGeneral(Tarea):
    #Constructor de la clase
    def __init__(self, nombre, descripcion, prioridad, estado, notas):
        super().__init__(nombre, descripcion, prioridad)
        self.estado= estado
        self.notas = notas

    #Muestra los detalles de la tarea
    def mostrarTarea(self):
        print(f"Nombre: {self.nombre}")
        print(f"Descripcion: {self.descripcion}")
        print(f"Prioridad: {self.prioridad}")
        print(f"Estado: {self.estado}")
        print(f"Notas: {self.notas}")
        print("..................") 
        print()

# ******** Defino subclase Tarea Con Fecha hereda de Tarea General ********
class TareaConFecha(TareaGeneral):
    #Constructor de la clase
    def __init__(self, nombre, descripcion, prioridad, estado, notas, fechaEntrega):
        super().__init__(nombre, descripcion, prioridad, estado, notas)
        self.fechaEntrega = fechaEntrega

    #Muestra los detalles de la tarea
    def mostrarTarea(self):
        print(f"Nombre: {self.nombre}")
        print(f"Descripcion: {self.descripcion}")
        print(f"Prioridad: {self.prioridad}")
        print(f"Estado: {self.estado}")
        print(f"Notas: {self.notas}")
        print(f"Fecha de entrega: {self.fechaEntrega}")
        print("..................")
        print()

#------------ Funciones -----------
    
    #Crea Tarea simple (solicita variables necesarias e instancia la tarea)
def crearTarea():
    nombre = input("Ingrese el nombre de la tarea: ")
    descripcion = input("Ingrese la descripcion de la tarea: ")
    prioridad = input("Ingrese la prioridad de la tarea: ")
    tarea = Tarea(nombre, descripcion, prioridad)#Instancia tarea
    return tarea

    #Crea Tarea General (solicita variables necesarias e instancia la tarea)
def crearTareaGeneral():
    nombre = input("Ingrese el nombre de la tarea: ")
    descripcion = input("Ingrese la descripcion de la tarea: ")
    prioridad = input("Ingrese la prioridad de la tarea: ")
    estado = input("Ingrese el estado de la tarea: ")
    notas = input("Ingrese una nota/recordatorio de la tarea: ")
    tarea_General = TareaGeneral(nombre, descripcion, prioridad, estado, notas) #Instancia tarea
    return tarea_General

   #Crea Tarea con fecha (solicita variables necesarias e instancia la tarea)
def crearTareaConFecha():
    nombre = input("Ingrese el nombre de la tarea: ")
    descripcion = input("Ingrese la descripcion de la tarea: ")
    prioridad = input("Ingrese la prioridad de la tarea: ")
    estado = input("Ingrese el estado de la tarea: ")
    notas = input("Ingrese una nota/recordatorio de la tarea: ")
    fechaEntrega = input("Ingrese una fecha de entrega de la tarea: ")
    tarea_ConFecha = TareaConFecha(nombre, descripcion, prioridad, estado, notas, fechaEntrega)#Instancia tarea
    return tarea_ConFecha

    #Elimina tarea de la lista buscando por su nombre
def eliminarTarea(ListaTareas):
    nombre = input("Ingrese el nombre de la tarea a eliminar: ")
    for tarea in ListaTareas: #Recorre la lista
        if tarea.nombre == nombre: #Compara el nombre
            ListaTareas.remove(tarea) #Remueve tarea
            print("¡¡¡ TAREA ELIMINADA !!!")
            print()

    #Muestra el listado de tareas guardadas
def mostrarListadoTareas(ListaTareas):
    print("---------- LISTADO DE TAREAS ----------")
    for tarea in ListaTareas: #Recorre la lista
        tarea.mostrarTarea() #Muestra tarea

            

#------ Programa Principal ------
ListaTareas = [] 

#Bucle para administrar las tareas hasta que desee salir
while True:
    print(" -------- Administrador de Tareas  --------")
    print("1. Crear tarea")
    print("2. Crear tarea general")
    print("3. Crear tarea con fecha")
    print("4. Eliminar tarea")
    print("5. Mostrar listado de tareas")
    print("6. Salir")
    opcion = int(input("Ingrese una opción: ")) #Solicito opcion al usuario
    print()

#Control de opciones con sus respectivas ordenes
    if(opcion == 1):
        ListaTareas.append(crearTarea())

    elif(opcion == 2):
        ListaTareas.append(crearTareaGeneral())

    elif(opcion == 3):
        ListaTareas.append(crearTareaConFecha())

    elif(opcion == 4):
        eliminarTarea(ListaTareas)

    elif(opcion == 5):
        mostrarListadoTareas(ListaTareas)

    elif(opcion == 6):
        break
    
    else:
        print("Opción no válida")





```
